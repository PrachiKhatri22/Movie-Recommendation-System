## Top 200 Movies based on IMDB and No. of Reviews

```python
sorted_movies = movies.sort_values(by=['imdb_score', 'num_user_for_reviews'], ascending=[False, False])

top_200_movies = sorted_movies.head(200)

```

```python
top_200_movies[['movie_title', 'imdb_score', 'num_user_for_reviews']]
```

```python
top_200_movies=top_200_movies.head(20)
# Create a scatter plot with IMDb score on the x-axis and number of user reviews on the y-axis
plt.figure(figsize=(12, 8))
sns.scatterplot(x='imdb_score', y='num_user_for_reviews', data=top_200_movies, hue='movie_title', palette='viridis', size='num_user_for_reviews', sizes=(20, 350))

# Set plot labels and title
plt.title('Top 200 Movies Based on IMDb Score and Number of Reviews')
plt.xlabel('IMDb Score')
plt.ylabel('Number of User Reviews')

# Display the plot
plt.show()
```

## Top 10 Directors

```python
average_scores = movies.groupby('director_name')['imdb_score'].mean()
```

```python
top_directors = average_scores.sort_values(ascending=False).head(10)
```

```python
top_directors
```

```python
top_directors_df = pd.DataFrame(
    {'DIRECTORS': top_directors.index,
     'Average_IMDb_Score': top_directors.values
    })
```

```python
top_directors_df
```

```python
plt.figure(figsize=(10, 8))


ax = sns.pointplot(x=top_directors_df['DIRECTORS'], y=top_directors_df['Average_IMDb_Score'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

## Recommending Movies Based on Actors

```python
def recommend_movies_by_actor(actor_name, movies):
    actor_movies = movies[movies['actor_1_name'] == actor_name].append(movies[movies['actor_2_name'] == actor_name]).append(movies[movies['actor_3_name'] == actor_name])
    recommended_movies = actor_movies.sort_values(by='imdb_score', ascending=False)
    recommended_movies = recommended_movies.head(5)
    return recommended_movies[['movie_title', 'imdb_score']]
```

```python
movies['actor_1_name']
```

```python
#actor_name = 'Johnny Depp'
actor_name = input("Enter actor name  : ")
recommended_movies = recommend_movies_by_actor(actor_name, movies)

print(f"Recommended movies for {actor_name}:")
recommended_movies
```
