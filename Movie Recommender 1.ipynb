## Top 200 Movies based on IMDB and No. of Reviews

```python
sorted_movies = movies.sort_values(by=['imdb_score', 'num_user_for_reviews'], ascending=[False, False])

top_200_movies = sorted_movies.head(200)
```

## Top 10 Directors

```python
average_scores = movies.groupby('director_name')['imdb_score'].mean()
average_scores
```

```python
top_directors = average_scores.sort_values(ascending=False).head(10)
```

## Recommending Movies Based on Actors

```python
def recommend_movies_by_actor(actor_name, movies_data):
    actor_movies = movies[movies['actor_1_name'] == actor_name].append(movies[movies['actor_2_name'] == actor_name]).append(movies[movies['actor_3_name'] == actor_name])
    recommended_movies = actor_movies.sort_values(by='imdb_score', ascending=False)
    recommended_movies = recommended_movies.head(5)
    return recommended_movies[['movie_title', 'imdb_score']]
```

```python
#actor_name = 'Christian Bale'
actor_name = input("Enter actor name  : ")
recommended_movies = recommend_movies_by_actor(actor_name, movies)

print(f"Recommended movies for {actor_name}:")
recommended_movies
```
